#Análise de Microdados do ENEM

```python
import pandas as pd
import matplotlib
```

```python
microdadosEnem = pd.read_csv('MICRODADOS_ENEM_2019.csv', sep=';', encoding='ISO-8859-1')
```

```python
microdadosEnem.head()
```

```python
microdadosEnem
```

```python
microdadosEnem.columns.values
```

```python
microdadosEnemSelecionado = microdadosEnem.filter(items = ['CO_MUNICIPIO_RESIDENCIA','NO_MUNICIPIO_RESIDENCIA',
                             'CO_UF_RESIDENCIA','SG_UF_RESIDENCIA','NU_NOTA_CN',
                             'NU_NOTA_CH','NU_NOTA_LC', 'NU_NOTA_MT','NU_NOTA_REDACAO'])
```

```python
microdadosEnemSelecionado
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_RESIDENCIA',
                                                            'NO_MUNICIPIO_RESIDENCIA',
                                                            'CO_UF_RESIDENCIA',
                                                            'SG_UF_RESIDENCIA']).mean()
```

```python
mediaEnemPorMunicipio
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionado.groupby(['CO_MUNICIPIO_RESIDENCIA',
                                                            'NO_MUNICIPIO_RESIDENCIA',
                                                            'CO_UF_RESIDENCIA',
                                                            'SG_UF_RESIDENCIA']).mean()
```

```python
mediaEnemPorMunicipio
```

```python
mediaEnemPorMunicipio = microdadosEnemSelecionado\
.groupby(['CO_MUNICIPIO_RESIDENCIA',
        'NO_MUNICIPIO_RESIDENCIA',
        'CO_UF_RESIDENCIA',
        'SG_UF_RESIDENCIA'],as_index = False).mean()
```

```python
mediaEnemPorMunicipio
```

```python
mediaEnemPorMunicipio.to_csv('MediaEnemPorMunicipio.csv')
```
