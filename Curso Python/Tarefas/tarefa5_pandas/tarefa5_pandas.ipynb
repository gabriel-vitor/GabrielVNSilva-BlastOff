```python
import pandas as pd
import numpy as np
```

```python
paises = pd.read_csv('paises.csv', sep=';',encoding='utf-8')
```

```python
paises.head()
```

Quais os países da OCEANIA;

```python
#a string ''OCEANIA'' possui espaços dentro dela, um total de 28 espaços vazios
#é possível utilizar a função strip() para remover esses espaços

oceania = paises.filter(items=['Country']).where(paises['Region'] == 'OCEANIA                            ')
oceania
```

```python
#para remover os valores nulos, usa a função dropna()
oceania = oceania.dropna()
oceania
```

Quantos países são da OCEANIA.

```python
quantidade = len(oceania)
print('A quantidade de países da Oceania é:', quantidade)
```

Encontre o nome e a região do país que possui a maior população;

```python
population = paises[['Country', 'Population']]
```

```python
population = population.sort_values(['Population'], ascending=False)
```

```python
most = population.loc[population['Population'].idxmax()]
most
```

Agrupe os países do dataset por Regiões. Em seguida, mostre a média de alfabetização (Literacy (%)) de cada região do planeta;

```python
regioes = paises.groupby('Region')
regioes['Literacy (%)'].mean()
```

Busque o nome de todos os países do dataset que não possuem costa marítima (Coastline (coast/área ratio) == 0) e guarde-os em um novo arquivo chamado noCoast.csv;

```python
no_coast = paises.filter(items=['Country']).where(paises['Coastline (coast/area ratio)']==0)
no_coast
```

```python
#remover valores nulos
no_coast = no_coast.dropna()
no_coast
```

```python
quant_nocoast = len(no_coast)
print('a quantidade de países sem costa é:', quant_nocoast)
```

```python
no_coast.to_csv('noCoast.csv')
```
